```python
def standard_deviation(original_data, mean_window_output):
    'This function calculates the standard deviation for each pixel according to its neighborhood.'
    step_1 = original_data - mean_window_output;
    step_2 = step_1**2;
    step_3 = uniform_filter(step_2, size=3, mode='constant')
    final = np.sqrt(step_3);
    return [final] #Doesn't work unless final is returned.    
```

```python
stdev = standard_deviation(a,b)
print (stdev)
```

```python
def standardised_anomaly(original_data, mean_grid, stdev_grid):
    #Calculates standardised anomalies.
    standard_anomaly = (original_data - mean_grid)/stdev_grid
    return [standard_anomaly]
    
```

```python
print (standardised_anomaly(a,b,stdev))
```

```python
#This is just showing that the ratio between the the population standard

import numpy as np
from random import randrange, uniform

arr1 = np.random.randint(10, size=10)

print (len(arr1)-1)

arr1x = sum(arr1)/10

arr1_anom = arr1 - arr1x

arr1_sqr = arr1_anom**2

print (arr1_sqr)

arr1_pop_var = sum(arr1_sqr)/ (len(arr1))

arr1_sample_var = sum(arr1_sqr)/ (len(arr1)-1)

arr1_stdev_pop = np.sqrt(arr1_pop_var)

arr1_stdev_sample = np.sqrt(arr1_sample_var)

print ('stdev population:', arr1_stdev_pop)
print ('stdev sample:', arr1_stdev_sample)


diff = arr1_stdev_pop / arr1_stdev_sample

print('diff: ', diff)
```
